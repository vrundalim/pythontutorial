```python
#creating the studentlist using dict
studentlist = {"id": pd.Series([101,102,103,104],index=['a','b','c','d']),"name": pd.Series(["asha","rekha","mahesh","rambha"],index=['a','b','c','d'])}
d  = pd.DataFrame(studentlist)
d

#select the  specific column of thee studentlist
d['id']

#to add the new column
d["eng"] = pd.Series([56,34,66,33],index=['a','b','c','d'])
d["math"] = pd.Series([54,78,55,67],index=['a','b','c','d'])
d["guj"] = pd.Series([45,56,78,55],index=['a','b','c','d'])
d["sci"] = pd.Series([47,55,66,77],index=['a','b','c','d'])
d["s.s"] = pd.Series([56,67,87,88],index=['a','b','c','d'])
d["total"] =  pd.Series(index=['a','b','c','d'])
d["total"] = d["eng"]+d["math"]+d["guj"]+d["sci"]+d["s.s"] 
d["per"] =  pd.Series(index=['a','b','c','d'])
d["per"] = d["total"]/4
d
#to delete a  column
# del d['per']
# d
# c= d.pop('eng')

```

```python
#to access the specific  row  by the index number
d.iloc[1]
#to access the multiple row
d.iloc[[0,1]]
#to access  the  row by  the range
d.iloc[0:3]
#to access the alternative  row
d.iloc[0: :2]
#to access the  single  column
d.iloc[:,4]
#to access the  multiple column bu range
d.iloc[:,0:8]
#to  select the  multiple column
d.iloc[:,[0,1,7,8]]
#condition
d.iloc[list(d["per"]>80)]
```

```python
#to access the  specfic row by the index of label
d.loc['b']
#to access the multiple row
d.loc[['a','b']]
#to  access the multiple row by  the  range
d.loc["a":"c"]
#to access the alternative  row
d.loc["a":"c":2]
#to  select the  multiple column
d.loc[:,["id","name","total","per"]]
#to access the  single  column
d.loc[:,"guj"]
#to access the  multiple column by  the  range
d.loc[:,"id":"per"]
#conditon
d.loc[d["per"]>80]
```
