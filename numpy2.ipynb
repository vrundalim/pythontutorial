```python
def genrate_time(in_sec):
   total_sec = in_sec*60
   for i in range(total_sec+1):
      sec =  i%60
      min = i//60
      yield f"{min:02} :{sec:02}"
for j in genrate_time(3):
    time.sleep(0.1)
    print(j)
```
