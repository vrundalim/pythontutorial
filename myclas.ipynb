```python
class myclass:
  def __init__(self):
      print("*******Student's Marksheet*******") 
  def Marks(sub,marks):
    if marks <= 100 and marks >=91:
        print(f"{sub} marks is  :{marks} and grade is A1")
    elif marks <= 90 and marks >=81:
        print(f"{sub} marks is  :{marks} and grade is A2")
    elif marks <= 80 and marks >=71:
        print(f"{sub} marks is  :{marks} and grade is B1")
    elif marks <= 70 and marks >=61:
        print(f"{sub} marks is  :{marks} and grade is B2")
    elif marks <= 60 and marks >=51:
        print(f"{sub} marks is  :{marks} and grade is C1")
    elif marks <= 50 and marks >=41:
        print(f"{sub} marks is  :{marks} and grade is C2")
    elif marks <= 40 and marks >=33:
        print(f"{sub} marks is  :{marks} and grade is D1")
    else:
        print("Fail")

cc = myclass()
myclass.Marks("eng",45)   
myclass.Marks("maths",35)  
myclass.Marks("sci",67)  
```
